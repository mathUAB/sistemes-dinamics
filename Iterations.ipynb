#### Iterations for 1-dimensional discrete systems

```python
def f(x):
    return 5-6./x
def iter1(f,n,x0): # f is the function of iteration, n the number of iterations and x0 the initial condition.
    aux = x0.n()
    L=[aux]
    for _ in range(n):
        aux=f(aux)
        L.append(aux)
    return L
```

```python
iter1(f,30,27)
```
